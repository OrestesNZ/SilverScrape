```python
import pandas as pd
import matplotlib.pyplot as plt
import quandl as quandl
import numpy as np
```

### Silver Bull Run 1971-74

#### Import data from Quandl, convert to csv

```python
data_71_74 = quandl.get("LBMA/SILVER", start_date="1971-01-01", end_date="1974-12-31", api_key='dTFXFSLUvt2MiM3bWUQd')
data_71_74['USD'].to_csv('silver_bulls_71_74.csv')
```

```python
silver_71_74 = pd.read_csv(r'C:\Users\Jayde\silver_bulls_71_74.csv')

silver_71 = silver_71_74[0:255]
silver_72 = silver_71_74[255:509].reset_index()
silver_73 = silver_71_74[509:762].reset_index()
silver_74 = silver_71_74[762:1020].reset_index()

silver_bull_71_74 = pd.concat([silver_71['USD'],silver_72['USD'],silver_73['USD'],silver_74['USD'],], axis=1)
silver_bull_71_74['Average'] = silver_bull_71_74.mean(axis=1)
silver_bull_71_74
```

```python
#plt.plot(silver_bull_71_74)
plt.plot(silver_bull_71_74['Average'])
```

### Silver Bull Run 76-80

```python

silver_bulls =  pd.read_csv(r'C:\Users\Jayde\SilverBulls.csv')
pd.set_option('display.max_rows', 260)

silver_bulls.head()
silver_bulls_usd = silver_bulls[['Date', 'USD']]

silver_bulls_1976 = silver_bulls_usd[0:254]
silver_bulls_1977 = silver_bulls_usd[254:506].reset_index()
silver_bulls_1978 = silver_bulls_usd.loc[506:758].reset_index()
silver_bulls_1979 = silver_bulls_usd.loc[758:1011].reset_index()
silver_bulls_1980 = silver_bulls_usd.loc[1011:].reset_index()


silver_bulls = pd.concat([silver_bulls_1976['USD'], silver_bulls_1977['USD'], silver_bulls_1978['USD'], silver_bulls_1979['USD'], silver_bulls_1980['USD']], axis=1)
silver_bulls['Average'] = silver_bulls.mean(numeric_only=True, axis=1)
silver_bulls['Log'] = np.log(silver_bulls['Average'])
silver_bulls

plt.figure(figsize=(20,10))
plt.plot(silver_bulls['Average'], label='Average Price Movement')
plt.plot(silver_bulls['Log'], label='Logarithmic Change')
plt.legend()
plt.ylabel('USD')
plt.xlabel('252 Trading Days of Y')
plt.title('Silver Bull Run 76-80')
plt.show()
```
